## Backprop on single neuron

```python
w = np.array([-3.0,-1.0,2.0])
bias = 1.0

inputs = np.array([1.0,0.8,2.6])

target_output = 0.0
learning_rate = 0.001

def relu(x):
    return np.maximum(0.0,x)

def relu_derivative(x):
    return 1 if relu(x)==x else 0

for iteration in range(20000):
    linear_output = np.dot(w,inputs) + bias
    output = relu(linear_output)
    loss = (output - target_output)**2

    dloss_doutput = 2 * (output-target_output)
    doutput_dlinear = relu_derivative(linear_output)
    dlinear_weights = inputs
    dlinear_dbias = 1.0

    dloss_dw = dloss_doutput*doutput_dlinear*dlinear_weights
    dloss_db = dloss_doutput*doutput_dlinear*dlinear_dbias

    w-= learning_rate*dloss_dw
    bias-= learning_rate*dloss_db

    print(f"Iteration: {iteration+1},loss: {loss}")

print(f"Final weight: {w}")
print(f"Final bias:{bias}")

```

## Back Prop a layer

```python
inputs = np.array([1,0.5,4,3])

w = np.array([
    [1,0.4,2,0.3],
    [0.4,0.8,4,2],
    [4,1,5,3]
])

b = np.array([0.4,5,3])

learning_rate = 0.001
target_output = 0

def relu(x):
    return np.maximum(x,0)

def relu_derivative(x):
    return (x > 0).astype(float)

for iteration in range(2000):
    z = np.dot(w,inputs) + b
    a = relu(z)

    loss = np.sum((a - target_output)**2)
    dL_da = 2*(a - target_output)
    #dy_da = np.ones_like(a) # y = sum(a-target) dy/da = 1
    da_dz = relu_derivative(z)

    dL_dz = dL_da*da_dz

    dz_dw = inputs

    dL_dw = np.outer(dL_dz,dz_dw)
    dL_db = dL_dz

    w-= learning_rate*dL_dw
    b-= learning_rate*dL_db

    print(f"Iteration {iteration+1}, Loss: {loss}")

print(f"Final weight: {w}")
print(f"Final bias: {b}")

```

## Loss function backprop

```python
class Loss:
  def caculate(self,output,y):
    sample_loss = self.forward(output,y)
    data_loss = np.mean(sample_loss)

    return data_loss
```

```python
class Loss_Cat_Cross_Entropy(Loss):
  def forward(self,y_pred,y_true):
    samples = len(y_pred)

    y_pred_clipped = np.clip(y_pred,1e-7,1-1e-7)

    if len(y_true.shape)==1:
      correct_confidences = y_pred_clipped[range(samples),y_true]
    elif len(y_true.shape)==2:
      correct_confidences = np.sum(
          y_pred*y_pred_clipped,
          axis=1
      )

    neg_log_likelihood = -np.log(correct_confidences)
    return neg_log_likelihood

  def backward(self,y_pred,y_true):
    samples = len(y_pred)

    labels = len(y_pred[0])

    if len(y_true.shape)==1:
      y_true = np.eye(labels)[y_true]

    self.dL_dy = -y_true/y_pred

    self.dL_dy = self.dL_dy/samples
```

## Softmax classifier - combined Softmax activation and cross-entropy loss for faster backward step


```python
class Activation_softmax:
    def forward(self,inputs):
        exp_vals = np.exp(inputs - np.max(inputs,axis=1,keepdims=True))
        prob = exp_vals/np.sum(exp_vals,axis=1,keepdims=True)
        self.output = prob
```

```python
class Activation_softmax_Loss_Cat_Cross:
  def __init__(self):
    self.activation = Activation_softmax()
    self.loss = Loss_Cat_Cross_Entropy()

  def forward(self,inputs,y_true):
    self.activation.forward(inputs)
    self.output = self.activation.output

    return self.loss.calculate(self.output,y_true)

  def backward(self,y_pred,y_true):
    samples = len(y_pred)

    if len(y_true.shape)==2:
      y_true = np.argmax(y_true,axis=1)

    self.dinputs = y_pred.copy()  # dinputs = derivative of the loss with respect to the inputs of that layer
    self.dinputs[range(samples),y_true]-=1

    self.dinputs = self.y/samples
```
